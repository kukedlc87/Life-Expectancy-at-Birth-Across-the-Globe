```python
df = pd.read_csv('/kaggle/input/life-expectancy-at-birth-across-the-globe/Life Expectancy at Birth.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

countries_of_interest = ['Afghanistan', 'Brazil', 'China', 'India', 'United States']

selected_countries = df[df['Country'].isin(countries_of_interest)]

life_expectancy_columns = df.columns[8:]

# Unstack the DataFrame
unstacked_df = selected_countries.melt(id_vars=['Country'], value_vars=life_expectancy_columns, 
                                       var_name='Year', value_name='Life Expectancy')

unstacked_df['Year'] = unstacked_df['Year'].str.extract('(\d+)').astype(int)

plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = unstacked_df[unstacked_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life Expectancy'], marker='o', label=country)

plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth')
plt.title('Life Expectancy at Birth by Country (1990-2021)')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the DataFrame (make sure to load your actual data file here)
# df = pd.read_csv('file_name.csv')

# Countries of interest for plotting
countries_of_interest = ['Rwanda', 'South Sudan', 'Timor-Leste', 'Malawi', 'Burundi', 'Hong Kong']

# Select the countries of interest
selected_countries = df[df['Country'].isin(countries_of_interest)]

# Columns for life expectancies (make sure these columns exist in your DataFrame)
life_expectancy_columns = df.columns[8:]  # Adjust the index according to your data

# Rearrange the DataFrame to have years as a single column and life expectancy values as values
unstacked_df = selected_countries.melt(id_vars=['Country'], value_vars=life_expectancy_columns, 
                                      var_name='Year', value_name='Life Expectancy')

# Convert the 'Year' column to numeric
unstacked_df['Year'] = unstacked_df['Year'].str.extract('(\d+)').astype(int)

# Create the plot
plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = unstacked_df[unstacked_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life Expectancy'], marker='o', label=country)

plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth')
plt.title('Life Expectancy at Birth by Country (1990-2021)')
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Select the life expectancy columns
life_expectancy_columns = df.columns[8:]  

# Calculate the standard deviation of life expectancy for each country
df['Variability'] = df[life_expectancy_columns].std(axis=1)

# Sort countries based on variability
countries_variability = df[['Country', 'Variability']].sort_values(by='Variability', ascending=False)

# Print the countries with the highest variability
print("Countries with Highest Variability in Life Expectancy:")
print(countries_variability.head(10))
```
